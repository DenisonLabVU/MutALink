```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv("./lineage1_haplotype_freq.txt", sep="\t", header=0)
df2=pd.read_csv("./lineage2_haplotype_freq.txt", sep="\t", header=0)
df3 = pd.read_csv("./lineage3_haplotype_freq.txt", sep="\t", header=0)
```

```python
df1
```

```python
df1_index = df1.set_index('Haplotype')
df2_index = df2.set_index('Haplotype')
df3_index = df3.set_index('Haplotype')
```

```python
sns.set()
fig, (ax1, ax2, ax3) = plt.subplots(ncols=3, sharex=True, sharey=False, figsize=(45,3))
im=sns.heatmap(df1_index, ax=ax1, cmap="RdPu", linewidth=0.5, linecolor="black", cbar=False, vmin=0, vmax=0.8)
ax1.set_xlabel('')
ax1.set_ylabel('')
ax1.set_title('Lineage 1', fontsize=20)
sns.heatmap(df2_index, ax=ax2, cmap="RdPu", linewidth=0.5, linecolor="black", cbar=False, vmin=0, vmax=0.8)
ax2.set_xlabel('')
ax2.set_ylabel('')
ax2.set_title('Lineage 2', fontsize=20)
sns.heatmap(df3_index, ax=ax3, cmap="RdPu", linewidth=0.5, linecolor="black", cbar=False, vmin=0, vmax=0.8)
ax3.set_xlabel('')
ax3.set_ylabel('')
ax3.set_title('Lineage 3', fontsize=20)
plt.setp(ax1.get_yticklabels(), rotation=0, fontsize=14)
plt.setp(ax2.get_yticklabels(), rotation=0, fontsize=14)
plt.setp(ax3.get_yticklabels(), rotation=0, fontsize=14)
plt.setp(ax1.get_xticklabels(), rotation=0, fontsize=16)
plt.setp(ax2.get_xticklabels(), rotation=0, fontsize=16)
plt.setp(ax3.get_xticklabels(), rotation=0, fontsize=16)
mappable = im.get_children()[0]
im.axhline(y = 0, color='k',linewidth = 5)
im.axhline(y = df1.shape[1], color = 'k',
            linewidth = 5)
im.axvline(x = 0, color = 'k',
            linewidth = 5)
im.axvline(x = df1.shape[0], 
            color = 'k', linewidth = 5)
#df2
im.axhline(y = 0, color='k',linewidth = 5)
im.axhline(y = df2.shape[1], color = 'k',
            linewidth = 5)
im.axvline(x = 0, color = 'k',
            linewidth = 5)
im.axvline(x = df2.shape[0], 
            color = 'k', linewidth = 5)
plt.colorbar(mappable, ax = [ax1,ax2,ax3], orientation = 'vertical', shrink=0.8, aspect=10, anchor=(-0.2, 0.5))
plt.text(3.1, 0.1, "Frequency", fontsize=16)
plt.show()
#plt.savefig("./SARS2_RDV_passage_heatmaps_filter.png", dpi=600)
```
